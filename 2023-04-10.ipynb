##### 3교시

```python
def min_max(x):
    return x.max()-x.min()
```

```python
df1.apply(min_max)
```

```python
# 그룹별로 만들어보자. 
grouped.agg(min_max)
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
df.groupby('key1').agg(peak_to_peak)
```

```python
df.groupby('key1')[['data1','data2']].apply(peak_to_peak)
```

```python
tips = pydataset.data('tips')
tips[:3]
```

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips[:3]
```

```python
list(tips.groupby(['day','smoker']))
```

```python
tips.groupby(['day','smoker']).mean()
```

```python
#  그룹별 칼럼마다 다른 함수 적용하기 
grouped = tips.groupby(['day','smoker'])
list(grouped['tip_pct'])
grouped['tip_pct'].agg('mean')
```

```python
grouped.agg({'tip':np.max,'size':'sum'})
```

```python
# 2가지 이상의 집계함수를 적용할 때 
list(grouped)
grouped['tip_pct'].mean()
grouped['tip_pct'].sum()
grouped['tip_pct'].max()
grouped['tip_pct'].agg(['mean','max','sum'])
```

```python
grouped['tip_pct'].agg([('평균','mean'),('합계','sum'),'max',('차이',peak_to_peak)])
```

```python
tips.sort_values('day')
```

```python
#  상위 5개의 데이터 프레임을 반환하시오.
def top(df,columns='tip_pct'):
    return df.sort_values(by='tip_pct',ascending = False)[:5]
```

```python
top(tips)
```
