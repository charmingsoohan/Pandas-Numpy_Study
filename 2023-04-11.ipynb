```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pydataset
```

```python
iris = pydataset.data('iris')
iris[:3]
```

```python
# 1.  Sepal.Length,Sepal.Width 를 Species별로 평균과 최대를 나타내시오.
iris.groupby('Species')[['Sepal.Length','Sepal.Width']].agg([('평균','mean'),('최대','max')])
```

```python
tips = pydataset.data('tips')
tips[:3]
```

```python
# 2.  tip_pct 칼럼을 만드시오.
tips['tip_pct'] = tips['tip']/(tips['total_bill']-tips['tip'])
tips
```

```python
# 3.  dat, smoker 로 그룹화를 하시오.
tips.groupby(['day','smoker'])
```

```python
# 4.  min, max, mean, sum 집계 함수를 적용하시오.
tips.groupby(['day','smoker']).agg(['min','max','mean','sum'])
```

```python
# 5.  다음 문자열 객체를 datetime 객체로 변환하시오.
from dateutil.parser import parse
parse('2023/04/10') 
```

```python
df =pd.DataFrame({'data1': np.random.randn(100), 'data2':np.random.randn(100)}) 
```

```python
# 6.  date1을 기준으로 4등분 하고 A,B,C,D 로 나눈후 grade 칼럼을 추가하시오.
df['grade'] = pd.cut(df.data1,4,labels=['A','B','C','D'])
df
```

```python
from pydataset import data
color = data("HairEyeColor")
color[:5] 
```

```python
# 7.  피벗 테이블을 사용해 hair,sex인덱스로 하고 eye를 칼럼으로 하는 데이터 프레임을 만드시오.
data = color.pivot_table(index=['Hair','Sex'],columns='Eye')
data
```

```python
# 8.  성별을 기준으로 
data.groupby('Sex').agg([('평균','mean'),('최대','max')])
```
