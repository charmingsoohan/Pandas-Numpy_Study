```python
#7번   users rating movies을 하나의 데이터 프레임으로 합쳐 주세요.
pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('users.dat.txt', sep='::',
                      header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ratings.dat.txt', sep='::',
                        header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat.txt', sep='::',
                       header=None, names=mnames, engine='python') 
```

```python
users
```

```python
data = pd.merge(pd.merge(users,ratings),movies)
data[:2]
```

```python
#8번  pivot_table을 사용하여 성별에 따른 각 영화의 평균 평점을 구하시오.
mean_ratings = data.pivot_table('rating',index='title',
                              columns='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
#9번  여성의 내림차순 , 남자는 오름차순으로 정렬하시오.
mean_ratings.sort_values(['F','M'],ascending=[False,True])
```

```python
years = range(1880,2011)
pieces = []
columns = ['name','sex','births']

for year in years:
    path = 'datasets/babynames/yob%d.txt'%year
    frame = pd.read_csv(path,names=columns)
    
    frame['year']=year
    pieces.append(frame)
    
names = pd.concat(pieces,ignore_index=True)
```

```python
names
```

```python
total_births = names.pivot_table('births',index='year',
                                columns='sex',aggfunc=sum)
total_births.tail()
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    group['prop'] = group.births / group.births.sum()
    return group
names = names.groupby(['year','sex']).apply(add_prop)
```

```python
names
```

```python
names.groupby(['year','sex']).prop.sum()
```
